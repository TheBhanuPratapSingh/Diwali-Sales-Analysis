# Exploratory Data Analysis

**Gender**

```python
df.columns
```

```python
ax = sns.countplot(x = "Gender", data = df, color = "r")
sns.set(rc={"figure.figsize":(3,3)})
for bars in ax.containers:
    ax.bar_label(bars)
```

**From above analysis it is clear that most of the buyers are Female**

```python
sales_gen = df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)
sns.barplot(x ="Gender", y = "Amount", data = sales_gen)
plt.show()
```

Above Analysis is showing that Amount spent by female on shoping is greater than Male

**Age**

```python
df.columns
```

```python
ax = sns.countplot(data = df, x = "Age Group", hue = "Gender")
for bars in ax.containers:
    ax.bar_label(bars)
sns.set(rc = {"figure.figsize":(8,6)})
```

```python
sales_age = df.groupby(["Age Group"], as_index = False)["Amount"].sum().sort_values(by = "Amount", ascending = False)
sns.barplot(x = "Age Group", y= "Amount", data = sales_age)
plt.show()
```

From Above Analysis Most of the buyers are Female and age between 26 to 35

**State**

```python
df.columns
```

```python
sales_state = df.groupby(["State"], as_index = False)["Orders"].sum().sort_values(by = "Orders", ascending = False).head(10)
sns.set(rc = {"figure.figsize":(18,5)})
sns.barplot(data = sales_state, x = "State", y = "Orders")
plt.show()
```

```python
sales_state = df.groupby(["State"], as_index = False)["Amount"].sum().sort_values(by = "Amount", ascending = False).head(10)
sns.set(rc = {"figure.figsize":(18,5)})
sns.barplot(data = sales_state, x = "State", y = "Amount")
plt.show()
```

These are the top 10 states from where the most of the buyers are and largest amount has been spent

**Marital Status**

```python
ax = sns.countplot(data =df, x = "Marital_Status")
sns.set(rc={"figure.figsize":(3,3)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(["Marital_Status","Gender"], as_index = False)["Amount"].sum().sort_values(by = "Amount", ascending = False)
sns.set(rc = {"figure.figsize":(3,4)})
sns.barplot(data = sales_state, x = "Marital_Status", y = "Amount", hue = "Gender")
plt.show()
```

Most of the buyers are married women

**Occupation**

```python
sns.set(rc = {"figure.figsize":(20,5)})
ax = sns.countplot(data = df, x = "Occupation")

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(["Occupation"], as_index = False)["Amount"].sum().sort_values(by = "Amount", ascending = False).head(10)
sns.set(rc = {"figure.figsize":(20,5)})
sns.barplot(data = sales_state, x = "Occupation", y = "Amount")
plt.show()
```

From Above analysis we can see that the top buyers Profession

**Product Category**

```python
sns.set(rc = {"figure.figsize":(25,10),"font.weight": "bold"})

plt.title("Count of Products in Each Category", fontweight="bold")
plt.xlabel("Product Category", fontweight="bold")
plt.ylabel("Count",fontweight="bold")

ax = sns.countplot(data = df, x = "Product_Category")
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha="right")

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(["Product_Category"], as_index = False)["Amount"].sum().sort_values(by = "Amount", ascending = False).head(10)
sns.set(rc = {"figure.figsize":(21,5),"font.weight": "bold"})
sns.barplot(data = sales_state, x = "Product_Category", y = "Amount")
plt.show()
```

```python
sales_state = df.groupby(["Product_ID"], as_index = False)["Amount"].sum().sort_values(by = "Amount", ascending = False).head(10)
sns.set(rc = {"figure.figsize":(20,5),"font.weight": "bold"})
sns.barplot(data = sales_state, x = "Product_ID", y = "Amount")
plt.show()
```

```python
sns.set(style="darkgrid", font_scale=1.5, rc={"font.weight": "bold"})
sns.set_context("talk",font_scale = .5)
fig1, ax1 = plt.subplots(figsize = (12,4))
df.groupby("Product_ID")["Orders"].sum().nlargest(10).sort_values(ascending = False).plot(kind = "bar")
plt.show()
```
